**Task 1.2: Data Preprocessing**:  
- Write a function `split_data(X, y, frac, seed)` that first shuffles your training data and then splits it into a training and a test set. `frac` determines the relative size of the test dataset, `seed` makes sure we get reproducible 
results.  
- Write a function `preprocess(X)` which zero-centers your data and sets variance to one (per-feature).

```python
def split_data(X, y, frac=0.3, max_samples=None, seed=None):
    if seed is not None:
        np.random.seed(seed)
    print(X.shape, y.shape)
    ### implement the function here
    #shuffle arrays
    idx = np.arange(X.shape[0])
    idx_shuffled = np.random.permutation(idx)
    X = X[idx_shuffled,:]
    y = y[idx_shuffled, :]

    #divide into test and train
    split = int(X.shape[0]*frac)
    X_test = X[:split,:]
    X_train = X[split:,:]
    y_test = y[:split,:]
    y_train = y[split:,:]

    return X_train, X_test, y_train, y_test

def preprocess(X):
    X = X - np.mean(X,axis=0)
    X = X / np.std(X, axis=0)
    return X
    
```

First, use Linear Discriminant Analysis to separate the classes. As discussed in the Bishop in pg. 186-189, we can find a weight vector $\vec{w}$ that performs a projection of the i-th input data point $\vec{x}_i$

```python
# Exercise 2.1.1
def compute_lda_weights(X, y):
    X1 = X[y[:,0],:]
    X2 = X[y[:,0]!=True,:]
    m1 = np.mean(X1, axis=0)
    m2 = np.mean(X2, axis=0)
    m1 = np.reshape(m1, (m1.shape[0],1))
    m2 = np.reshape(m2, (m2.shape[0],1))
    cov = ((X1 - m1.T).T @ (X1 - m1.T)) + ((X2 - m2.T).T @ (X2 - m2.T))
    w = np.linalg.inv(cov) @ (m2 - m1)
    w = np.reshape(w, (w.shape[0], 1)) #w has no shape
    c = w.T @ (0.5*(m1 + m2))
    return w, c

def plot(w_lda):
    plt.stem(w_lda.flatten(), use_line_collection=True)
    plt.title('Computed LDA weights')
    plt.ylabel('weight')
    plt.xlabel('predictors')
```

```python
X_train, X_test, y_train, y_test = split_data(X_all, t_all, seed=1)
```

```python
w_lda, c = compute_lda_weights(X_train, y_train)
w_lda = np.reshape(w_lda, (w_lda.shape[0],1))
plot(w_lda)
```

```python
def plot_sep(a, b, c, ax, s):
    ax.set_title(s)
    _ = ax.hist(a,label='class 0',alpha=0.5)
    _ = ax.hist(b,label='class 1',alpha=0.5)
    _ = ax.axvline(c[0,0], c='red')
    ax.legend()

fig, ax = plt.subplots(1,2)
a = X_train[y_train[:,0]] @ w_lda
b = X_train[y_train[:,0]!=True] @ w_lda
plot_sep(a, b, c, ax[0], 'Training data projection')

a = X_test[y_test[:,0]] @ w_lda
b = X_test[y_test[:,0]!=True] @ w_lda
plot_sep(a, b, c, ax[1], 'Test data projection')
```
